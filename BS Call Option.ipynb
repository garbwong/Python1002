```python
from scipy import stats

# input parameters
S0 = 105.00 # stock price
K = 100.00 # strike price
T = 1. # option maturity
r = 0.05 # constant interest rate
sigma = 0.25 # constant volatility of diffusion

```

# Black Scholes Call Option Model

```python
from IPython.display import Image
Image(filename='BlackScholes.png')
```

```python
# Analytical Formula
def BSM_call_value(S0, K, T, r, sigma):
    d1 = (math.log(S0 / K) + (r + 0.5 * sigma ** 2) * T) / (sigma * math.sqrt(T))
    d2 = d1 - sigma * math.sqrt(T) 
    call_value = (S0 * stats.norm.cdf(d1, 0.0, 1.0) - K * math.exp(-r * T) * stats.norm.cdf(d2, 0.0, 1.0))
    return call_value
# Output
print("Value of European call option is %8.5f" % BSM_call_value(S0, K, T, r, sigma))
```
