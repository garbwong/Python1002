```python
from scipy import zeros, sqrt, shape, exp, mean
import scipy as sp
S0 = 105 # stock price at time zero
X= 100 # exercise price
T =1 # years
r =0.05 # risk-free rate
sigma = 0.25 # volatility (annual)
```

```python
from IPython.display import Image
Image(filename='simulation.png')
```

```python
n_steps=1000. # number of steps
sp.random.seed(1000) # set the seed
n_simulation = 10000 # number of simulation
dt =T/n_steps

call = zeros([n_simulation], dtype=float)
x = range(0, int(n_steps), 1)
for j in range(0, n_simulation):
    sT=S0
    total = 0
    for i in x[:-1]:
        e=sp.random.normal()
        sT*=exp((r-0.5*sigma*sigma)*dt+sigma*e*sqrt(dt))
    call[j]=max(sT-X,0)
call_price=mean(call)*exp(-r*T)

print('call price = ', round(call_price,4))
```

```python
# Call price districbution
plt.hist(call*exp(-r*T),bins=100)
plt.xlabel('Call price')
plt.ylabel('frequency')
plt.grid(True)
```
