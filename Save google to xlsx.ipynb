# Save Google data to xlsx

```python
from pandas_datareader import data, wb
import datetime as dt
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
start = dt.datetime(2013,1,1)
end = dt.datetime(2017,4,20)
#Search stock code and data
data = data.DataReader('AAPL','google', start, end)
```

```python
data.plot()
```

```python
data.tail()
```

```python
data['Close'].plot()
```

```python
data.to_excel('Stock_Price_APPL.xlsx')
```
